# Отток клиентов

Из «Бета-Банка» стали уходить клиенты. Каждый месяц. Немного, но заметно. Банковские маркетологи посчитали: сохранять текущих клиентов дешевле, чем привлекать новых.

Нужно спрогнозировать, уйдёт клиент из банка в ближайшее время или нет. Вам предоставлены исторические данные о поведении клиентов и расторжении договоров с банком. 

Постройте модель с предельно большим значением *F1*-меры. Чтобы сдать проект успешно, нужно довести метрику до 0.59. Проверьте *F1*-меру на тестовой выборке самостоятельно.

Дополнительно измеряйте *AUC-ROC*, сравнивайте её значение с *F1*-мерой.

Источник данных: [https://www.kaggle.com/barelydedicated/bank-customer-churn-modeling](https://www.kaggle.com/barelydedicated/bank-customer-churn-modeling)

**Цель исследования:** Заказчиком выступает банк (Бета-Банк), который желает получить систему, способную прогнозировать уйдёт клиент из банка в ближайшее время или нет.

**Задачи исследования:** Необходимо построить классификтор, способный выдать прогноз на предмет того уйдет клиент из банка или нет. Критерий эффективности - значение F1 меры на уровне не менее 0.59.

- RowNumber — индекс строки в данных

- EstimatedSalary — предполагаемая зарплата

**Целевой признак**

Проверим гипотезу о равенстве среднего возраста в двух классах:
- Н_0 - средний возраст у клиентов ушедших из банка равен среднему возрасту клиентов, оставшихся в банке.
- Н_1 - средний возраст у клиентов ушедших из банка не равен среднему возрасту клиентов, оставшихся в банке.

```python
# Гипотезу о равенстве средних значений между классами будем проверять с помощью T-test 
# Создадим простую функцию, к которой будем обращаться в дальнейшем
# Параметр equal_var укажем как False, учитывая, что мы не можем однозначно сказать, что дисперсии выборок равны, 
# а также мы знаем о дисбалансе выборок (выборки не равны между собой)

def get_ttest_value (target):
    results = st.ttest_ind(
    data[data['Exited'] == 0][target],
    data[data['Exited'] == 1][target],
    equal_var = False
    )

    alpha = 0.05

    print(results.pvalue)

    if (results.pvalue < alpha):
        print("Отвергаем нулевую гипотезу")
    else:
        print("Не получилось отвергнуть нулевую гипотезу")
```

```python
get_ttest_value('Age')
```

Учитывая, что гипотеза о равенстве среднего возраста между клиентами ушедшими из банка и оставшимися не подтвердилась, можно сделать вывод о том, что возраст клиета влияет на вероятность ухода из банка.

### Обучение моделей при дисбалансе классов

Проведем обучение случайного леса.Подбор гиперпараметров будем осуществлять в цикле


```python
def get_best_model_rf_by_f1(X_train, Y_train, X_valid, Y_valid, weight=None):

    best_model = None
    best_f1 = 0
    
    for est in np.arange(2, 41).astype(int):
        for depth in np.arange(1, 36).astype(int):
            model = RandomForestClassifier(random_state=12345,  n_estimators=est, max_depth=depth, class_weight=weight)
            model.fit(X_train, Y_train)
            prediction = model.predict(X_valid)
            result_f1 = f1_score(Y_valid, prediction)
            if result_f1 > best_f1:
                best_f1 = result_f1
                best_model = model
            
    return best_f1, best_model     
```

```python
# Получаем лучшую оценку F1 меры на валидационной выборке и модель
base_rf_f1, base_rf = get_best_model_rf_by_f1(features_train, target_train, features_valid, target_valid)
# CСчитаем значение ROC-AUC для лучшей модели по оценке F1 на валидационной выборке
base_rf_auc_roc = roc_auc_score(target_valid,base_rf.predict_proba(features_valid)[:, 1])

print('F1 мера для модели случайного леса на валидационной выборке = {:.4f}'.format(base_rf_f1))
print('ROC-AUC для модели случайного леса на валидационной выборке = {:.4f}'.format(base_rf_auc_roc))
```

```python
result_f1 = [base_rf_f1]
result_roc_auc = [base_rf_auc_roc]

index_list = ['RF']
columns_name  = ['F1_score', 'ROC-AUC']
result_base_model = pd.DataFrame(list(zip(result_f1, result_roc_auc)), 
                                 index=index_list, 
                                 columns=columns_name)
result_base_model
```

На несбалансированных данных метрика ROC-AUC для модели случайного леса на уровне 0.82 показывает, что модель отвечает лучше, чем случайным образом. В целом модель случайного леса демонстрирует необходимый уровень F1 на несбалансированных данных. Следующим шагом поработаем с устранением дисбаланса.

### Downsampling

#### Уменьшение обучающей выборки

```python
def downsample(features, target, fraction):
    
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]
    
    features_downsampled = pd.concat([features_zeros.sample(frac=fraction, random_state=12345)] + [features_ones])
    target_downsampled = pd.concat([target_zeros.sample(frac=fraction, random_state=12345)] + [target_ones])    
    
    features_downsampled, target_downsampled = shuffle(features_downsampled, target_downsampled, random_state=12345)    
    return features_downsampled, target_downsampled
```

### Upsampling

#### Увеличение обучающей выборки

```python
def upsample(features, target, repeat):
    
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]
    
    features_upsampled = pd.concat([features_zeros] + [features_ones] * repeat)
    target_upsampled = pd.concat([target_zeros] + [target_ones] * repeat)    
    
    features_upsampled, target_upsampled = shuffle(features_upsampled, target_upsampled, random_state=12345)   
    
    return features_upsampled, target_upsampled
```

Как видно из таблицы выше максимальное значение F1 меры и ROC-AUC у модели случайного леса с балансировкой классов. Попробуем изменить порог классификации. Оптимальный порог будем подбирать путем максимизации F1 меры.

```python
# Напишем функцию которая на вход прнимает массимы из признаков и целевых значений, 
# модель и ее наименование (для красивого вывода на экран)
# Функция будет печать лушчий найденный порог классификации для F1 меры

def get_optinal_threshold(features, target, model, model_name):
    
    probabilities_one_test = model.predict_proba(features)[:, 1]

    best_f1_score = 0
    best_threshold = 0
    best_predicted_test = 0

    for threshold in np.arange(0.1, 1, 0.01):

        predicted = probabilities_one_test > threshold 
        f1 = f1_score(target, predicted)

        if f1 > best_f1_score:
            best_f1_score = f1
            best_threshold = threshold
            
    print('Оптимальный порог классификации при котором F1 принимает максимальное значение для модели', model_name)
    print('Порог = {:.2f}, значение F1 = {:.4f}'
          .format(best_threshold, best_f1_score))
    print('--------------------------------------------------------------------------------------------')
    
    return best_f1_score, best_threshold
```

```python
models = [base_rf, bw_rf, rus_rf, ds_rf, us_rf, smote_rf]
models_name = ['RF', 'RF BW', 'RF RUS', 'RF DS', 'RF US', 'RF SMOTE']

for i in range(len(models)):
        get_optinal_threshold(features_valid, target_valid, models[i], models_name[i])
```

Было рассмотренно несколько методов борьбы с дисбалансом, каждый из них улучшал результаты по сравнениию с оценкой на не сбалансированных данных. При этом  методики балансирования классов и повышения размероности дали резульаты лучше, чем сокращение размерности. По результатам данного этапа отобрана лучшая модель (по максимальному значению F1 меры, которая будет проверена на тестовой выборке).
